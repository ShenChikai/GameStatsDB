# Steam Game Data Processing

```python
steamG = pd.read_csv('steam_games.csv')
steamG
```

```python
ngDict = {}
stDict = {}

for nidx, nName in enumerate(newGamedf.Name):
    for sidx, sName in enumerate(steamG.name):
        if nName == sName:
            ngDict[nName] = nidx
            stDict[sName] = sidx

print(len(ngDict))
print(len(stDict))
```

```python
for idx, row in newGamedf.iterrows():
    if row.Name in stDict and row.Platform == 'PC':
        print(row)
        newGamedf.loc[idx,'Platform'] = 'Steam'
```

```python
# from math import isnan
# free = ['Free', 'Demo Middle-earth™: Shadow of War™', 'Play for Free!', 'Shadow of the Tomb Raider Free Trial', 'FINAL FANTASY XV WINDOWS EDITION PLAYABLE DEMO', \
#             'UNO', 'Demo', 'CrossCode Demo', 'EmergeNYC Demo', 'SpeedRunners Party Mode', ]
for idx, row in newGamedf.iterrows():
    if row.Name in stDict:
        if steamG.loc[stDict[row.Name],'original_price'][0] == '$':
            if float(steamG.loc[stDict[row.Name],'original_price'][1:]) != 0.0:
                newGamedf.loc[idx, 'IsF2P'] = False
            else:
                newGamedf.loc[idx, 'IsF2P'] = True
        else:
            newGamedf.loc[idx, 'IsF2P'] = True
    else:
        newGamedf.loc[idx, 'IsF2P'] = False
newGamedf
```

# GameHasGenre Relation Table

```python
GameHasGenre = newGamedf[['GID', 'Genre']].drop_duplicates()
GameHasGenre = GameHasGenre.set_index('GID')
GameHasGenre
```

```python
for idx, row in GameHasGenre.iterrows():
    if len(row.Genre.split(',')) != 1:
        newGenre = row.Genre.replace(',', '/')
        newGenre.strip()
        newGenre.lstrip()
        newGenre.rstrip()
        GameHasGenre.loc[idx, 'Genre'] = newGenre

GameHasGenre 
```

```python
GameHasGenre
```

```python
GameHasGenre = GameHasGenre.reset_index()
j = GameHasGenre.shape[0]

for idx, row in GameHasGenre.iterrows():
    g = row.Genre.split('/')
    if len(g) != 1:
        GameHasGenre.loc[idx, 'Genre'] = g[0].strip().rstrip().lstrip()
        gid = row.GID
        for othergenre in g[1:]:
            GameHasGenre.loc[j] = [gid, othergenre.strip().rstrip().lstrip()]
            j+=1
GameHasGenre
```

## Stock Table

```python
tickers_yf = [yf.Ticker(x) for x in tickers]
tickers_yf[0].info.keys()
```

```python
stockinfo = defaultdict(list)
for idx, tk in enumerate(tickers):
    try:
        stockinfo[tk].extend([tickers_yf[idx].info['country'], tickers_yf[idx].info['exchange'], tickers_yf[idx].info['fiftyTwoWeekHigh'], tickers_yf[idx].info['fiftyTwoWeekLow'], tickers_yf[idx].info['market']])
    except:
        pass
```

```python
i = 0
for ticker, info in stockinfo.items():
    if i == 0:
        c, e, h, l, m = info
        stockdf = pd.DataFrame({'Ticker': ticker, 'Country':c, 'Exchange':e, 'FiftyTwoWeekHigh':h, 'FiftyTwoWeekLow':l, 'Market':m}, index=[i])
    else:
        if info:
            res = [ticker]
            res.extend(info)
            stockdf.loc[i,:] = res
    i += 1

stockdf
```

```python
stockdf.to_csv('clean/stock.csv', index=False)
```
